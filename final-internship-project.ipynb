```python
#Most played venues
#plotting
plt.figure(figsize=(16,8))
sns.countplot(x='venue',data=com_df ,order=com_df['venue'].value_counts().index)
plt.title('Venue which hosted most number of IPL matches', fontsize=15)
plt.xlabel('Venue', fontsize=15)
plt.xticks(rotation=90,fontsize=15)
plt.ylabel('Number of matches', fontsize=15)
```

```python
# Top umpires

umpires = pd.concat([com_df['umpire1'], com_df['umpire2']])

# Get the value counts of umpires
umpire_counts = umpires.value_counts().head(10)
colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'lightblue', 'lightcoral', 'yellow', 'lightgrey', 'lightseagreen', 'salmon']

# Plotting
plt.figure(figsize=(10, 6))
umpire_counts.plot(kind='bar',color=colors)
plt.title('Top 10 Umpires Based on Number of Matches')
plt.xlabel('Umpire')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**S Ravi** umpired most of the matches.

```python
#Top batsmen
top_batsmen = com_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'lightblue', 'lightcoral', 'yellow', 'lightgrey', 'lightseagreen', 'salmon']
# Plotting
plt.figure(figsize=(10, 6))
top_batsmen.plot(kind='bar', color=colors)
plt.title('Top 10 Batsmen by Total Runs Scored')
plt.xlabel('Batsmen')
plt.ylabel('Total Runs Scored')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**V kohli** is the batsmen who scored most runs in IPL.

```python
#T op bowlers
top_bowlers = com_df[~com_df['player_dismissed'].isnull()].groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10)

# Define colors for bars
colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'lightblue', 'lightcoral', 'yellow', 'lightgrey', 'lightseagreen', 'salmon']

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette=colors)  # Horizontal bar plot with Seaborn
plt.title('Top 10 Bowlers by Total Wickets Taken')
plt.xlabel('Total Wickets Taken')
plt.ylabel('Bowler')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Grid along x-axis
plt.show()
```

**L S Malinga** took more than 175 wickets through out IPL .

**toss impact on match outcomes**

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='toss_decision', data=com_df, palette='Set2')
plt.title('Frequency of Toss Decisions in IPL')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

Feilding has the most frequncy in toss decisions.

```python
toss_outcome = com_df.groupby(['toss_decision', 'winner']).size()
print("Impact of toss decisions on match outcomes:")
print(toss_outcome)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=toss_outcome.index.get_level_values('winner'), y=toss_outcome.values, hue=toss_outcome.index.get_level_values('toss_decision'))
plt.title('Impact of Toss Decisions on Match Outcomes')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend(title='Toss Decision')
plt.show()
```

**Chennai Super Kings** won 52 matches when they chose to bat first and 48 matches when they chose to field first.
**Delhi Capitals** won 31 matches when they chose to bat first and 46 matches when they chose to field first.

```python
batting_first_wins = com_df[com_df['toss_decision'] == 'bat']['winner'].value_counts()
batting_second_wins = com_df[com_df['toss_decision'] == 'field']['winner'].value_counts()
print("Influence of batting first vs. batting second:")
print("Wins after batting first:")
print(batting_first_wins)
print("\nWins after batting second:")
print(batting_second_wins)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=batting_first_wins.index, y=batting_first_wins.values, color='skyblue', label='Batting First')
sns.barplot(x=batting_second_wins.index, y=batting_second_wins.values, color='orange', label='Batting Second')
plt.title('Influence of Batting First vs. Batting Second on Match Outcomes')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Teams like Mumbai Indians performed better when they choose to bat later.

```python
total_batting_first_wins = batting_first_wins.sum()
total_batting_second_wins = batting_second_wins.sum()

# Create labels for the pie chart
labels = ['Batting First', 'Batting Second']

# Create data for the pie chart
sizes = [total_batting_first_wins, total_batting_second_wins]

# Define colors for the pie chart
colors = ['skyblue', 'orange']

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Wins - Batting First vs. Batting Second')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

Hence **Batting second** gives **more winning percentage**.
